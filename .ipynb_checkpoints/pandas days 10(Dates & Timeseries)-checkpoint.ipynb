```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read csv
df=pd.read_csv('ETH_1h.csv')
```

```python
df.head()
```

```python
df.info() #dataset information
df.describe() #statistical sumary
```

```python
# Converting date into datetime
df['Date']=pd.to_datetime(df['Date'],format='%Y-%m-%d %I-%p')
```

```python
df['Date']
```

```python
type(df.index)
```

```python
#Feature Engineering(Time-based features)
# Extracting Year, Month,Day from date
df['Year']=df.index.year
df['Month']=df.index.month
df['Day']=df.index.day
df['Hour']=df.index.hour
df['Weekday']=df.index.day_name()
df.head()
```

```python
#Filter data(Example:year 2017)
df_2017=df[df.index.year==2017]
df_2017.head()
```

```python
# Filter using date range
# Filter between two dates
df_2017 = df.loc['2017-01-01':'2017-12-31']

df_2017.head()
```

```python
#Resampling(Time aggregation)-Hourly-daily

# Daily average Close price
daily_close = df['Close'].resample('D').mean()

daily_close.head()
```

```python
# Monthly volume
# Monthly traded volume (Month End)
monthly_volume = df['Volume'].resample('ME').sum()

monthly_volume.head()

```

```python
#Monthly OHLC(financial data)
# Monthly Open, High, Low, Close
monthly_ohlc = df['Close'].resample('ME').ohlc()

monthly_ohlc.head()
```

```python
# Rolling window(Moving average)
# 24-hour moving average
df['MA_24'] = df['Close'].rolling(window=24).mean()

df[['Close', 'MA_24']].head(30)
```

```python
#Lag features(Important for ML)
# Previous hour close prices
df['Close_lag_1'] = df['Close'].shift(1)
df['Close_lag_2'] = df['Close'].shift(2)

df.head()
```

```python
# Check missing values
df.isna().sum()
```

```python
# Forward fill missing values
df.ffill(inplace=True)
```

```python
df.head(4)
```

```python
# Group by year
# Average Close price per year
yearly_avg = df.groupby(df.index.year)['Close'].mean()
yearly_avg
```
