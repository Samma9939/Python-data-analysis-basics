```python
import pandas as pd
import numpy as np
```

```python
people={
    'first':['samma','soni','sageena','musarrat',np.nan,None,'NA'],
    'last':['khatoon','bibi','begam','fatma',np.nan,np.nan,'Missing'],
    'email':['sammak@gmail.com','sonib@gmail.com','sageenabegam@gmail.com',None,np.nan,'anonymous@gmail.com','NA'],
    'age':['20','30','45','19',None,None,'Missing']
    }
```

```python
df=pd.DataFrame(people)
df.replace('NA',np.nan,inplace=True)
df.replace('Missing',np.nan,inplace=True)
```

```python
df
```

```python
df.dropna()
```

```python
df.dropna(axis='index',how='any')
```

```python
df.dropna(axis='index',how='all')
```

```python
df.dropna(axis='columns',how='all')
```

```python
df.dropna(axis='columns',how='any')
```

```python
df.dropna(axis='index',how='any',subset=['email'])
```

```python
df.dropna(axis='index',how='all',subset=['email'])
```

```python
df.dropna(axis='index',how='all',subset=['last','email'])
```
